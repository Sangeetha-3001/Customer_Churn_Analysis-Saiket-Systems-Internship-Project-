```python
import pandas as pd
df=pd.read_excel(r"C:\Users\HP\Documents\Customer_Churn_Analysis Tasks\Customer_Churn_Dataset.xlsx")
```

```python
print(df.columns.tolist())
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
import matplotlib.pyplot as plt

df.select_dtypes(include="number").hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

```python
Churn_percentage = df['Churn'].value_counts(normalize=True)*100
print(Churn_percentage)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
plt. figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

*Customers who Churn are fewer than customers who stay
*Most customers are retained by the company.
*Numerical features show different distribution across customers.
*The correltion matrix helps identify relationships between numerical variables.
*These insights can be used to improve customer retention strategies
