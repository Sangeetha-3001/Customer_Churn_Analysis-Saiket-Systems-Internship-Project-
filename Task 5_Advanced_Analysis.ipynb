```python
df["Tenure_Group"] = pd.cut(
    x=df["tenure"],
    bins=[-1, 12, 36, 72],
    labels=["0-12 Months", "13-36 Months", "37+ Months"]
)

print(df[["tenure", "Tenure_Group"]].head())
```

```python
tenure_analysis = df.groupby("Tenure_Group").agg({
    "MonthlyCharges": "mean",
    "TotalCharges": "mean"
})

print(tenure_analysis)
```

```python
gender_churn = pd.crosstab(df["gender"], df["Churn"])
print(gender_churn)
```

```python
senior_churn = pd.crosstab(df["SeniorCitizen"], df["Churn"])
print(senior_churn)
```

```python
payment_churn = pd.crosstab(df["PaymentMethod"], df["Churn"])
print(payment_churn)
```

```python
contract_churn = pd.crosstab(df["Contract"], df["Churn"])
print(contract_churn)
```

```python
tenure_analysis["MonthlyCharges"].plot(kind="bar", figsize=(6,4))
plt.title("Average Monthly Charges by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Average Monthly Charges")
plt.xticks(rotation=0)
plt.show()
```

```python
gender_churn.plot(kind="bar", figsize=(6,4))
plt.title("Churn by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

```python
senior_churn.plot(kind="bar", figsize=(6,4))
plt.title("Churn by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

```python
payment_churn.plot(kind="bar", figsize=(8,5))
plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

```python
contract_churn.plot(kind="bar", figsize=(6,4))
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

Customers with shorter tenure and month-to-month contracts show higher churn. Churn patterns also vary by payment method and customer demographics. These insights can help the company improve customer retention by targeting high-risk customer groups with suitable offers and long-term contract plans.

# Business Recommendations

1. Focus on customers with 0–12 months tenure by offering welcome discounts and loyalty benefits.
2. Encourage customers to switch from month-to-month contracts to long-term contracts.
3. Improve customer support for high-risk customer groups.
4. Offer personalized plans based on customer payment methods and usage patterns.
5. Monitor churn regularly to improve customer retention.
